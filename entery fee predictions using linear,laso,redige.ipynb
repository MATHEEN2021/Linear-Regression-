```python
import numpy as np 
import pandas as pd 
df = pd.read_csv("E:\PYTHON WITH MACHINE LEARNING\Top Indian Places to Visit.csv")
print(df.to_string()) 
```

```python
df.nunique()
```

```python
col = ['Type','Google review rating','Number of google review in lakhs','Entrance Fee in INR']
df1 = df[col]
print(df1.to_string())
```

```python
df1 = df1[df1['Entrance Fee in INR'] != 0]
print(df1.to_string())
```

```python
df1 = pd.get_dummies(df1,dtype = 'int')
```

```python
X = df1.drop(columns = ['Entrance Fee in INR'])
Y = df1['Entrance Fee in INR']
print(X,Y.to_string())
```

```python
from sklearn.linear_model import LinearRegression
modela = LinearRegression()
```

```python
modela.fit(X,Y)
```

```python
modela.score(X,Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33)
modelb = LinearRegression()
modelb.fit(X,Y)
```

```python
modelb.score(X_train,Y_train)
```

```python
modelb.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Lasso
model3= Lasso(alpha=50,max_iter=100,tol=0.3)
model3.fit(X_train,Y_train)
```

```python
model3.score(X_train,Y_train)
```

```python
model3.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Ridge
model4= Ridge(alpha=50,max_iter=100,tol=0.3)
model4.fit(X_train,Y_train)
```

```python
model4.score(X_train,Y_train)
```

```python
model4.score(X_test,Y_test)
```
